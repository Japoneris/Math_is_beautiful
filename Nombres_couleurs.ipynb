# Primes 

```python
n = 100

series = capot.prime(n)
print("{} to be shown".format(len(series)))

for i in series:
    fig, ax = plt.subplots(figsize=(15,15))
    mat = capot.generate(i)
    ax.imshow(mat)
    ax.set_title("Size: {}".format(i))
    
```
